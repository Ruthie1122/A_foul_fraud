```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up visualization parameters
sns.set(style="whitegrid")

# 1. Bar Chart: Distribution of Fraudulent vs Non-Fraudulent Transactions
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=Dataframe, palette=["#3498db", "#e74c3c"])
plt.title("Distribution of Fraud vs Non-Fraud Transactions")
plt.xlabel("Transaction Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Number of Transactions")
plt.xticks([0, 1], ['Non-Fraudulent', 'Fraudulent'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=Dataframe, palette=["#3498db", "#e74c3c"])
plt.ylim(0, 1000)  # Limiting the y-axis to better visualize the differences
plt.title("Transaction Amounts by Class (Fraud vs Non-Fraud)")
plt.xlabel("Transaction Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Transaction Amount ($)")
plt.show()
```

```python
corr = Dataframe.corr()
plt.figure(figsize=(14, 12))
sns.heatmap(corr, cmap="coolwarm", annot=False, fmt=".2f")
plt.title("Correlation Heatmap of All Features")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(Dataframe, x = Dataframe.index, y = Dataframe['Amount'], hue= Dataframe['Class'])
#x = df_creditcard.index : chaque point du scatter plot sera positionné horizontalement en fonction des valeurs de l’index du DataFrame
# Donc on aura 284807 point sous l'axe X
plt.title('Distribution of Amount Values')
plt.xlabel('Transactions')
plt.ylabel('Amount of transaction in EUR')
plt.show()
```

```python
Dataframe_fraud = Dataframe.loc[Dataframe['Class'] == 1]

plt.figure(figsize=(10,5))
sns.scatterplot(Dataframe, x = Dataframe.index, y = Dataframe['Amount'], hue= Dataframe['Class'])
plt.title('Distribution of Amount Values')
plt.xlabel('Transactions')
plt.ylabel('Amount of transaction in EUR')

sns.scatterplot(Dataframe_fraud, x = Dataframe_fraud.index, y = Dataframe_fraud['Amount'], color = 'Orange')
plt.show()
```
